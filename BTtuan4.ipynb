```python
#Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n."
import math

def largest_prime_factor(n):
    max_prime = -1

    while n % 2 == 0:
        max_prime = 2
        n //= 2

    for i in range(3, int(math.sqrt(n)) + 1, 2):
        while n % i == 0:
            max_prime = i
            n //= i

    if n > 2:
        max_prime = n

    return max_prime

try:
    n = int(input("Nhập vào số nguyên dương n (n >= 2): "))
    if n < 2:
        print("Vui lòng nhập n >= 2")
    else:
        result = largest_prime_factor(n)
        print(f"Ước số nguyên tố lớn nhất của {n} là: {result}")
except ValueError:
    print("Vui lòng nhập một số nguyên hợp lệ.")
```

```python
#ReverseAddPalindrome] Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số đối xứng (palindrome). In ra số bước và giá trị palindrome thu được.
def reverse_number(num):
    return int(str(num)[::-1])

def find_palindrome_steps(n):
    steps = 0
    while str(n) != str(n)[::-1]:
        n = n + reverse_number(n)
        steps += 1
    return steps, n

n = int(input())
steps, palindrome = find_palindrome_steps(n)
print(steps, palindrome)
```

```python
#[PerfectSquaresDistinctDigits] Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu cách). Nếu không có, in ra (no number). 
import math

def has_distinct_digits(num):
    s = str(num)
    return len(s) == len(set(s))

def solve():
    try:
        line = input().strip()
        if not line:
            return
        n = int(line)
    except EOFError:
        return
    except ValueError:
        return

    results = []
    
    i = 1
    while i * i <= n:
        square = i * i
        if has_distinct_digits(square):
            results.append(str(square))
        i += 1
    
    if results:
        print(" ".join(results))
    else:
        print("(no number)")

if __name__ == "__main__":
    solve()
```

```python
#Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số nhỏ nhất." 
def collatz_length(m, memo):
    original = m
    length = 1 
    while m != 1:
        if m in memo:
            length += memo[m] - 1
            break
        if m % 2 == 0:
            m //= 2
        else:
            m = 3 * m + 1
        length += 1
    memo[original] = length
    return length

def main():
    n = int(input().strip())
    memo = {1: 1}
    max_len = 0
    result_num = 1
    
    for i in range(1, n + 1):
        length = collatz_length(i, memo)
        if length > max_len:
            max_len = length
            result_num = i
    
    print(result_num, max_len)

if __name__ == "__main__":
    main()
```

```python
#Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật: Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại. Ví dụ 220 và 284 là một cặp số thân thiết 220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55, 110. Tổng của 11 ước này cho chúng ta 284 284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng vừa đúng bằng 220 Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không. Nếu có, in ra màn hình "true", ngược lại, in ra "false" 
def sum_proper_divisors(num):
    total = 0
    for i in range(1, int(num**0.5) + 1):
        if num % i == 0:
            total += i
            if i*i != num and i != 1:
                total += num // i
    return total

a = int(input())
b = int(input())

if sum_proper_divisors(a) == b and sum_proper_divisors(b) == a:
    print("true")
else:
    print("false")
```

```python
#Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình tìm ước chung lớn nhất của hai số m và n.
m = int(input())
n = int(input())

def tim_ucln(a, b):
    while b != 0:
        a, b = b, a % b
    return a
ucln = tim_ucln(m, n)
print(f"{ucln}")
```
