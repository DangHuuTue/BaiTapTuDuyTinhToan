```python
#Tính tam giác: Nhập vào ba số a,b,c bất kỳ. Kiểm tra xem nó có thể là độ dài ba cạnh hay không...
import math
a = float(input())
b = float(input())
c = float(input())   
if a + b > c and a + c > b and b + c > a:
    chu_vi = a + b + c
    p = chu_vi / 2
    dien_tich = math.sqrt(p * (p - a) * (p - b) * (p - c))        
    loai = "Tam giác thường"
    if a == b == c:
        loai = "Tam giác đều"
    elif a == b or b == c or a == c:
        if (a**2 + b**2 == round(c**2, 5)) or (a**2 + c**2 == round(b**2, 5)) or (b**2 + c**2 == round(a**2, 5)):
            loai = "Tam giác vuông cân"
        else:
            loai = "Tam giác cân"
    elif (a**2 + b**2 == round(c**2, 5)) or (a**2 + c**2 == round(b**2, 5)) or (b**2 + c**2 == round(a**2, 5)):
        loai = "Tam giác vuông"           
    print(f"Đây là {loai}. Chu vi: {chu_vi}, Diện tích: {dien_tich:.2f}")
else:
    print("Khong la ba canh cua mot tam giac")
```

```python
#Viết chương trình giải phương trình bậc 2 (ax^2 + bx + c = 0)
import math
a = float(input())
b = float(input())
c = float(input())
if a == 0:
    if b == 0: print("Vô số nghiệm" if c == 0 else "Vô nghiệm")
    else: print(f"Nghiệm x = {-c/b}")
else:
    delta = b**2 - 4*a*c
    if delta < 0:
        print("Phương trình vô nghiệm")
    elif delta == 0:
        print(f"Nghiệm kép x = {-b/(2*a)}")
    else:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        print(f"Có 2 nghiệm phân biệt: x1 = {x1}, x2 = {x2}")
```

```python
#Giải hệ phương trình tuyến tính (ax + by = m và cx + dy = n)
print("ax + by = m và cx + dy = n")
a, b, m = float(input("a: ")), float(input("b: ")), float(input("m: "))
c, d, n = float(input("c: ")), float(input("d: ")), float(input("n: "))
    
D = a * d - b * c
Dx = m * d - b * n
Dy = a * n - m * c
    
if D != 0:
    print(f"Hệ có nghiệm duy nhất: x = {Dx/D}, y = {Dy/D}")
else:
    if Dx == 0 and Dy == 0:
        print("Hệ có vô số nghiệm")
    else:
        print("Hệ vô nghiệm")
```

```python
#Nhập tâm và bán kính đường tròn. Kiểm tra điểm A(x, y) có thuộc đường tròn không?
import math
x_tam = float(input())
y_tam = float(input())
r = float(input())
xa = float(input())
ya = float(input())
    
khoang_cach = math.sqrt((xa - x_tam)**2 + (ya - y_tam)**2)
    
if khoang_cach == r:
    print("Điểm A nằm trên đường tròn")
elif khoang_cach < r:
        print("Điểm A nằm trong đường tròn")
else:
        print("Điểm A nằm ngoài đường tròn")
```
