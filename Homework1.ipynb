```python
# Bài 1: Nhập vào số nguyên n, in ra màn hình số gấp đôi của n
n = int(input())
print(n * 2)
```

```python
# Bài 2: Tính diện tích còn lại của khu đất (Diện tích HCN - Diện tích hình tròn)
a = float(input())
b = float(input())
pi = 3.14
r = a / 2
s_hcn = a * b
s_tron = pi * (r ** 2)
s_con_lai = s_hcn - s_tron
print(f"{round(s_con_lai, 2)}")
```

```python
# Bài 3: Nhập kí tự c, hoa in thường, thường in hoa
c = input()
if c.isupper():
    print(c.lower())
elif c.islower():
    print(c.upper())
else:
    print(c)
```

```python
# Bài 4: Kiểm tra c có phải là kí tự alphabet không
c = input()
if c.isalpha():
    print(f"'{c}' là kí tự alphabet")
else:
    print(f"'{c}' không phải là kí tự alphabet")
```

```python
# Bài 5: Nhập chữ hoa A, in ra chữ cái thường liền trước chữ cái thường tương ứng của A. Xét TH c = 'A'
c = input()
char_code = ord(c.lower())
if c == 'A':
    print("Trường hợp đặc biệt: Không có chữ cái thường liền trước 'a'")
else:
    print(chr(char_code - 1))
```

```python
# Bài 6: Kiểm tra 3 cạnh tam giác, tính diện tích bằng công thức Heron
import math
a = float(input())
b = float(input())
c = float(input())
if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    s = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print(f"Diện tích: {round(s, 1)}")
else:
    print("Khong phai 3 canh cua tam giac")
```

```python
# Bài 7: Nhập chuỗi độ dài >= 20, in ra chữ cái thứ 5 và thứ 9
s = input()
if len(s) >= 20:
    print(f"Kí tự thứ 5: {s[4]}, Kí tự thứ 9: {s[8]}")
else:
    print("Chuỗi quá ngắn!")
```

```python
# Bài 8: Tính tiền điện theo bậc thang và VAT 8%
ten = input()
truoc = int(input())
nay = int(input())
kwh = nay - truoc
tien = 0
if kwh <= 50:
    tien = kwh * 1984
elif kwh <= 100:
    tien = 50 * 1984 + (kwh - 50) * 2050
elif kwh <= 200:
    tien = 50 * 1984 + 50 * 2050 + (kwh - 100) * 2380
elif kwh <= 300:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + (kwh - 200) * 2998
elif kwh <= 400:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (kwh - 300) * 3350
else:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (kwh - 400) * 3460
tong_tien = tien * 1.08
print(f"Ho va ten: {ten}")
print(f"Tien phai tra la: {int(tong_tien)}")
```
