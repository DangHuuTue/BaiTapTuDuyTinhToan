```python
# Bài 1: In các số dương lẻ bé hơn 100
for i in range(1, 100, 2):
    print(i, end=" ")
```

```python
# Bài 2: In các số dương chẵn chia hết cho 3 bé hơn 100
for i in range(1, 100):
    if i % 2 == 0 and i % 3 == 0:
        print(i, end=" ")
```

```python
# Bài 3: Nhập vào số nguyên dương a, in ra bảng cửu chương của a
a = int(input())
for i in range(1, 11):
    print(f"{a} x {i} = {a * i}")
```

```python
# Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím
h = int(input())
for i in range(1, h + 1):
    for j in range(1, 2 * h):
        if j == h - i + 1 or j == h + i - 1 or i == h:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

```python
# Bài 5: Nhập vào số nguyên dương n. Tính S = 1 + 2 + 3 + 4 + ... + n
n = int(input())
s = 0
for i in range(1, n + 1):
    s += i
print("Tổng S =", s)
```

```python
# Bài 6: Nhập vào số nguyên dương a, in ra tất cả các ước dương của a
a = int(input())
print(f"Các ước của {a} là:", end=" ")
for i in range(1, a + 1):
    if a % i == 0:
        print(i, end=" ")
```

```python
# Bài 7: Nhập vào số nguyên dương a, đếm số ước dương của a
a = int(input())
dem = 0
for i in range(1, a + 1):
    if a % i == 0:
        dem += 1
print(f"Số lượng ước của {a} là: {dem}")
```

```python
# Bài 8: Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b
a = int(input())
b = int(input())
print(f"Ước chung của {a} và {b} là:", end=" ")
limit = a if a < b else b
for i in range(1, limit + 1):
    if a % i == 0 and b % i == 0:
        print(i, end=" ")
```

```python
# Bài 9: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay không?
a = int(input())
ip = False
for i in range(a + 1):
    if i * i == a:
        ip = True
        break
if ip:
    print(f"{a} là số chính phương")
else:
    print(f"{a} không phải là số chính phương")
```

```python
# Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không.
a = int(input())
tong_uoc = 0
for i in range(1, a):
    if a % i == 0:
        tong_uoc += i
if tong_uoc == a:
    print(f"{a} là số hoàn hảo")
else:
    print(f"{a} không phải là số hoàn hảo")
```

```python
# Bài 11: Nhập vào số nguyên a, nếu nhập số âm thì yêu cầu nhập lại...
while True:
    a = int(input())
    if a > 0:
        print("Yes")
        break
```

```python
# Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + ... + k. Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n
n = int(input())
sk = 0
k = 0
while sk + (k + 1) < n:
    k += 1
    sk += k
print(f"{sk}")
```
