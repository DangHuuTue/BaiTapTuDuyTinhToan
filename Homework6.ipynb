```python
# Nhập vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải
n = int(input())
m = int(input())
matrix = []
for i in range(n):
    row = list(map(int, input(f"{i+1}").split()))
    matrix.append(row)
for row in matrix:
    for val in row:
        print(f"{val:>4}", end="")
    print()
```

```python
# Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và đường chéo phụ của mảng
n = int(input())
matrix = []
for i in range(n):
    row = list(map(int, input(f"{i+1}").split()))
    matrix.append(row)
main_diagonal = [matrix[i][i] for i in range(n)]
anti_diagonal = [matrix[i][n - 1 - i] for i in range(n)]
print("Đường chéo chính:", *main_diagonal)
print("Đường chéo phụ:", *anti_diagonal)
```

```python
# Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất trong list
def max_vl(lst):
    if not lst:
        return None
    max_val = max(lst)
    return lst.index(max_val)
lst = list(map(int, input().split()))
max_vl(lst)
```

```python
# Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về -1
def vt(lst, k):
    try:
        return lst.index(k)
    except ValueError:
        return -1
lst = list(map(int, input().split()))
k = int(input())  
vt(lst, k)  
```

```python
# Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện
input_str = input().split()
danh_muc = {}
for i, word in enumerate(input_str):
    if word not in danh_muc:
        danh_muc[word] = []
    danh_muc[word].append(i)
result = {k: tuple(v) for k, v in danh_muc.items()}
print(result)
```

```python
# Viết hàm xóa đi các số < x có trong mảng
def xoa_nho_hon_x(lst, x):
    return [i for i in lst if i >= x]
lst = list(map(int, input().split()))
x = int(input())
xoa_nho_hon_x(lst, x)
```

```python
# Viết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 0 2 3 2 4 9
def day_so_0_ve_dau(lst):
    z = [i for i in lst if i == 0]
    o = [i for i in lst if i != 0]
    return z + o
print(day_so_0_ve_dau([0, 2, 3, 2, 4, 0, 0, 0, 9]))
```

```python
# Cho dãy chỉ gồm N giá trị từ 0 - 9. Kiểm tra xem giá trị nào không có trong mảng
def tim_so_thieu(lst):
    all_digits = set(range(10))
    present_digits = set(lst)
    missing = all_digits - present_digits
    return sorted(list(missing))
lst = list(map(int, input().split()))
print(tim_so_thieu(lst))
```

```python
# Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2 phần tử đầu và cuối
def tong_cuc_dai_dia_phuong(lst):
    tong = 0
    for i in range(1, len(lst) - 1):
        if lst[i] > lst[i-1] and lst[i] > lst[i+1]:
            tong += lst[i]
    return tong
lst = list(map(int, input().split()))
tong_cuc_dai_dia_phuong(lst)
```

```python
# Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước). nếu không có in ra -1
def tim_vi_tri_nho_hon_x(lst, x):
    indices = [i for i, val in enumerate(lst) if val < x]
    if not indices:
        print(-1)
    else:
        print(*indices)
lst = list(map(int, input().split()))
x = int(input())
tim_vi_tri_nho_hon_x(lst, x)
```

```python
# Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm thực sự, hay vô danh (không tăng, không giảm)
def kiem_tra_day_so(lst):
    if len(lst) < 2:
        return "Vô danh"
    is_increasing = all(lst[i] < lst[i+1] for i in range(len(lst)-1))
    is_decreasing = all(lst[i] > lst[i+1] for i in range(len(lst)-1))
    if is_increasing:
        return "Tăng thực sự"
    elif is_decreasing:
        return "Giảm thực sự"
    else:
        return "Vô danh"
lst = list(map(int, input().split()))
kiem_tra_day_so(lst)
```
