```python
#Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả. Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3] 
l = map(int, input().split())
rs = []
for i in l:
    if i not in rs:
        rs.append(i)
print(rs)
```

```python
#Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu tới vị trí i), giữ nguyên thứ tự
l = map(int, input().split())
rs = 0
n = []
for i in l:
    rs += i
    n.append(rs)
print(n)
```

```python
#Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple sau khi xoay  trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp xếp.)
t = tuple(map(int, input().split()))
k = int(input())
n = len(t)
k = k % n
kq = t[k:] + t[:k]
print(kq)
```

```python
#Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key lặp lại, thêm vào list của key đó
pairs = input().split()
rs = {}
for p in pairs:
    k,v = p.split(':')
    if k not in rs:
        rs[k] = []
    rs[k].append(v)
rs
```

```python
#Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 'zeros': ...} (đếm số >0, <0, =0) 
l = list(map(int, input().split()))
a = sum(1 for x in l if x > 0)
b = sum(1 for x in l if x < 0)
c = sum(1 for x in l if x == 0)
q = {}
q['positives'] = a
q['negatives'] = b
q['zero'] = c
q
```

```python
# Viết chương trình nhập vào một list các chuỗi, sau đó đếm số lần xuất hiện của từng chuỗi và lưu vào dictionary.
words = input().split()
d = {}
for w in words:
    d[w] = d.get(w, 0) + 1
print(d)
```

```python
# Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của key.
import ast
def get_dict():
    line = input()
    try:
        return ast.literal_eval(line)
    except:
        parts = line.split()
        d = {}
        for i in range(0, len(parts), 2):
            d[parts[i]] = int(parts[i+1])
        return d
dict1 = get_dict()
dict2 = get_dict()
res = dict1.copy()
for k, v in dict2.items():
    res[k] = res.get(k, 0) + v
sorted_res = {k: res[k] for k in sorted(res.keys())}
print(sorted_res)
```

```python
# Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.
nums = list(map(int, input().split()))
k = int(input())
res = []
for i in range(len(nums)):
    for j in range(i + 1, len(nums)):
        if nums[i] + nums[j] == k:
            res.append((nums[i], nums[j]))
res.sort()
print(res)
```

```python
# [Tách số chẵn và lẻ từ tuple] Nhập vào một tuple chứa các số nguyên. Hãy tạo ra hai tuple mới: một chứa tất cả các số chẵn và một chứa tất cả các số lẻ từ tuple ban đầu, giữ nguyên thứ tự tương đối. In ra hai tuple này trên hai dòng riêng biệt.
tup = tuple(map(int, input().split()))
even = tuple(x for x in tup if x % 2 == 0)
odd = tuple(x for x in tup if x % 2 != 0)
print(even)
print(odd)
```

```python
# [Tìm phần tử xuất hiện nhiều nhất] Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ nhất trong số đó.
nums = list(map(int, input().split()))
if not nums:
    exit()
counts = {}
for x in nums:
    counts[x] = counts.get(x, 0) + 1
max_freq = max(counts.values())
candidates = [k for k, v in counts.items() if v == max_freq]
print(min(candidates))
```

```python
# [Đảo ngược key và value của dictionary] Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo một dictionary mới bằng cách đảo ngược key và value của dictionary ban đầu. In ra dictionary mới.
parts = input().split()
d = {}
for i in range(0, len(parts), 2):
    d[parts[i]] = parts[i+1]
res = {v: k for k, v in d.items()}
print(res)
```

```python
# [Giao của hai danh sách] Nhập vào hai dòng, mỗi dòng là một danh sách các số nguyên. Hãy tìm các phần tử chung (giao) của hai danh sách này và in ra một list mới chứa các phần tử chung đó, không trùng lặp.
list1 = list(map(int, input().split()))
list2 = list(map(int, input().split()))
res = []
seen = set()
set2 = set(list2)
for x in list1:
    if x in set2 and x not in seen:
        res.append(x)
        seen.add(x)
print(res)
```

```python
# [Lọc dictionary theo giá trị] Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp key- value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary mới.
parts = input().split()
d = {}
for i in range(0, len(parts), 2):
    d[parts[i]] = int(parts[i+1])
k = int(input())
res = {key: val for key, val in d.items() if val > k}
print(res)
```
