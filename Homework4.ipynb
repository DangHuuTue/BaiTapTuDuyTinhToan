```python
#Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b [cite: 2]
a = int(input())
b = int(input())
limit = min(a, b)
print(f"Các ước chung của {a} và {b}:")
for i in range(1, limit + 1):
    if a % i == 0 and b % i == 0:
        print(i, end=" ")
print()
```

```python
#Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không [cite: 2]
n = int(input())
temp_n = n
is_power_of_two = True
if n <= 0:
    is_power_of_two = False
else:
    while temp_n > 1:
        if temp_n % 2 != 0:
            is_power_of_two = False
            break
        temp_n //= 2

if is_power_of_two:
    print(f"{n} là số dạng 2^k")
else:
    print(f"{n} không phải là số dạng 2^k")
```

```python
#Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ [cite: 2]
n = int(input())
chan = 0
le = 0
for chu_so in str(n):
    if int(chu_so) % 2 == 0:
        chan += 1
    else:
        le += 1
print(f"Chữ số chẵn: {chan}, Chữ số lẻ: {le}")
```

```python
#Nhập vào A, tìm n nhỏ nhất sao cho 1+1/2+1/3+1/4+...+1/n > A [cite: 2]
A = float(input())
s = 0
n = 0
while s <= A:
    n += 1
    s += 1/n
print(f"Số n nhỏ nhất là: {n}")
```

```python
#Nhập n. Cho S(k)=1+2+3+...+k. Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n [cite: 2]
n = int(input())
s_k = 0
k = 0
while s_k + (k + 1) < n:
    k += 1
    s_k += k
print(f"Số k tìm được là: {k} (với S({k}) = {s_k})")
```

```python
#Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập [cite: 2]
max_val = None
min_val = None
while True:
    num = int(input())
    if num == -1:
        break
    if max_val is None or num > max_val:
        max_val = num
    if min_val is None or num < min_val:
        min_val = num

if max_val is not None:
    print(f"Lớn nhất: {max_val}, Nhỏ nhất: {min_val}")
```

```python
#[Fibonaci1] Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A [cite: 2]
A = int(input())
f1, f2 = 1, 1
while f2 <= A:
    temp = f1 + f2
    f1 = f2
    f2 = temp
print(f"Số Fibonacci lớn nhất <= {A} là: {f1}")
```

```python
#Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng). Dùng cách: for trên index; KHÔNG dùng hàm split(), len() [cite: 2]
s = input()
count_words = 0
length = 0
for char in s:
    length += 1

in_word = False
for i in range(length):
    if s[i] != " ":
        if not in_word:
            count_words += 1
            in_word = True
    else:
        in_word = False
print(f"Số lượng từ: {count_words}")
```

```python
#Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi. Dùng cách: for trên index; KHÔNG dùng hàm split() [cite: 2]
s = input()
length = 0
for char in s: length += 1

first_word = ""
start_copy = False
for i in range(length):
    if s[i] != " ":
        start_copy = True
        first_word += s[i]
    elif start_copy:
        break
print(f"Từ đầu tiên: {first_word}")
```

```python
#Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành thường và thường thành hoa. [cite: 2]
s = input()
result = ""
for char in s:
    if 'a' <= char <= 'z':
        result += char.upper()
    elif 'A' <= char <= 'Z':
        result += char.lower()
    else:
        result += char
print("Kết quả:", result)
```

```python
#Nhập vào một chuỗi có dạng 3 số nguyên, mỗi số nguyên cách nhau một dấu phẩy, hãy tính tổng 3 số nguyên đó. VD: Nhập: 3, 12, 15 -> Tổng: 30 [cite: 2]
s = input()
parts = s.split(",")
tong = 0
for p in parts:
    tong += int(p.strip())
print("Tổng:", tong)
```

```python
#Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in thường, bao nhiêu ký tự số [cite: 2]
s = input()
hoa = 0
thuong = 0
so = 0
for char in s:
    if char.isupper():
        hoa += 1
    elif char.islower():
        thuong += 1
    elif char.isdigit():
        so += 1
print(f"In hoa: {hoa}, In thường: {thuong}, Số: {so}")
```
