```python
# Nhập vào một số nguyên dương có 3 chữ số. Tìm và in ra màn hình SỐ NGUYÊN đảo ngược của n. 
# Không dùng kiểu chuỗi. Chú ý: kiểm tra chỉ xử lý các số nguyên dương có đúng 3 chữ số. Các số khác không xử lý.
n = int(input())
if 100 <= n <= 999:
    dv = n % 10
    chuc = (n // 10) % 10
    tram = n // 100
    dao_nguoc = dv * 100 + chuc * 10 + tram
    print(dao_nguoc)
```

```python
#Hoán đổi hai số không sd biến tạm thời. Sử dụng phép toán XOR trên bit để hoán đổi giá trị của hai biến.
a = int(input())
b = int(input())
a = a ^ b
b = a ^ b
a = a ^ b
print(a, b)
```

```python
#Kiểm tra xem một số có phải là lũy thừa của 2 hay không. Sử dụng các toán tử bitwise.
# Gợi ý: Đối với lũy thừa của 2, n & (n - 1) == 0
n = int(input())
if n > 0 and (n & (n - 1)) == 0:
    print("Yes")
else:
    print("No")
```

```python
#[Even and Odd] Nhập vào một số nguyên dương x, kiểm tra xem x là số chẵn hay lẻ. 
# Nếu x là số chẵn, in ra "Even". Nếu x là số lẻ, in ra "Odd".
x = int(input())
if x % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
#[Check negative numbers] Nhập vào hai số nguyên a, b. 
# Nếu cả hai số a, b đều âm thì in ra "Yes", ngược lại in ra "No".
a = int(input())
b = int(input())
if a < 0 and b < 0:
    print("Yes")
else:
    print("No")#
```

```python
#[Longer string] Nhập vào hai chuỗi a, b gồm các chữ cái thường. 
# In ra "True" nếu chuỗi a dài hơn chuỗi b. Ngược lại in ra "False".
a = input()
b = input()
if len(a) > len(b):
    print("True")
else:
    print("False")
```

```python
#[Triangle] Nhập vào 3 số nguyên dương a, b, c. Kiểm tra xem a, b, c có cấu thành độ dài cạnh của 1 tam giác được không.
a = int(input())
b = int(input())
c = int(input())
if a + b > c and a + c > b and b + c > a:
    print("Yes")
else:
    print("No")
```

```python
#[Largest number] Nhập vào 4 số nguyên. In ra màn hình số nguyên lớn nhất trong 4 số.
nums = [int(input()) for _ in range(4)]
print(max(nums))
```

```python
#[Phân loại tam giác] Nhập vào 3 số nguyên dương a,b,c. Kiểm tra tam giác hợp lệ. 
# Phân loại "Tam giác đều", "Tam giác cân", hay "Tam giác thường". Nếu không hợp lệ in "Không phải tam giác".
a = int(input())
b = int(input())
c = int(input())
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print("Tam giác đều")
    elif a == b or b == c or a == c:
        print("Tam giác cân")
    else:
        print("Tam giác thường")
else:
    print("Không phải tam giác")
```

```python
#[Kiểm tra năm nhuận] Nhập vào số nguyên dương n là số năm. 
# Năm nhuận chia hết cho 4 nhưng không chia hết cho 100, hoặc chia hết cho 400.
n = int(input())
if (n % 4 == 0 and n % 100 != 0) or (n % 400 == 0):
    print("Yes")
else:
    print("No")
```

```python
#[Giải phương trình bậc nhất] Nhập vào hai số thực a và b của phương trình ax + b = 0.
# In nghiệm làm tròn đến số thập phân thứ 2. In "Vô nghiệm" hoặc "Vô số nghiệm" tương ứng.
a = float(input())
b = float(input())
if a == 0:
    if b == 0:
        print("Vô số nghiệm")
    else:
        print("Vô nghiệm")
else:
    x = -b / a
    print(f"{x:.2f}")
```

```python
#[Xếp loại học lực] Nhập vào điểm trung bình. 
# >= 8.0: Gioi, >= 6.5: Khá, >= 5.0: Trung bình, < 5.0: Yếu.
dtb = float(input())
if dtb >= 8.0:
    print("Gioi")
elif dtb >= 6.5:
    print("Khá")
elif dtb >= 5.0:
    print("Trung bình")
else:
    print("Yếu")
```

```python
#[Rounding] Nhập vào một số thực: in ra kết quả làm tròn lên, tròn xuống, làm tròn tới số nguyên gần nhất 
# (không được sử dụng hàm sẵn có của python). Các kết quả cách nhau một dấu cách.
n = float(input())

if n >= 0:
    floor_val = int(n)
else:
    floor_val = int(n) if n == int(n) else int(n) - 1

if n >= 0:
    ceil_val = int(n) if n == int(n) else int(n) + 1
else:
    ceil_val = int(n)

if n >= 0:
    round_val = int(n + 0.5)
else:
    round_val = int(n - 0.5)

print(f"{ceil_val} {floor_val} {round_val}")
```
