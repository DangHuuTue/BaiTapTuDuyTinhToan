```python
# Viết hàm kiểm tra một số có phải số Armstrong hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x
def is_armstrong(x):
    s = str(x)
    n = len(s)
    total = sum(int(digit) ** n for digit in s)
    return total == x
num = int(input())
print(is_armstrong(num))
```

```python
# Viết hàm kiểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x
import math
def is_perfect_square(x):
    if x < 0: return False
    root = math.isqrt(x)
    return root * root == x
num = int(input())
print(is_perfect_square(num))
```

```python
# Viết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.
def is_palindrome(x):
    s = str(x)
    return s == s[::-1]
num = int(input())
print(is_palindrome(num))
```

```python
# Viết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.(Giả định: Số kỳ diệu là số có tổng các ước thực sự bằng chính nó - số hoàn hảo)
def is_magic_number(x):
    if x <= 0: return False
    total = sum(i for i in range(1, x) if x % i == 0)
    return total == x
num = int(input())
print(is_magic_number(num))
```

```python
# Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.
def detect_number_type(s):
    try:
        if float(s).is_integer() and '.' not in str(s):
            return 'int'
        float(s)
        return 'float'
    except ValueError:
        return 'neither'
val = input()
print(detect_number_type(val))
```

```python
# Viết hàm is_one_product_of_others(a, b, c) trả về True nếu một trong ba số bằng tích của hai số còn lại
def is_one_product_of_others(a, b, c):
    return a == b * c or b == a * c or c == a * b
a = float(input())
b = float(input())
c = float(input())
print(is_one_product_of_others(a, b, c))
```

```python
# Viết hàm sum_halves(s) nhận chuỗi 6 chữ số và trả về tổng của hai số 3 chữ số đầu và 3 chữ số cuối.
def sum_halves(s):
    if len(s) == 6 and s.isdigit():
        return int(s[:3]) + int(s[3:])
    return False
s_input = input()
print(sum_halves(s_input))
```

```python
# Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình ax^2 + bx + c = 0.
def quadratic_roots_count(a, b, c):
    if a == 0:
        if b == 0:
            return -1 if c == 0 else 0
        return 1
    delta = b**2 - 4*a*c
    if delta > 0: return 2
    elif delta == 0: return 1
    else: return 0
a = float(input())
b = float(input())
c = float(input())
print(quadratic_roots_count(a, b, c))
```

```python
# Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.
def are_collinear(x1, y1, x2, y2, x3, y3):
    return (y2 - y1) * (x3 - x2) == (y3 - y2) * (x2 - x1)
coords = list(map(float, input().split()))
print(are_collinear(*coords))
```

```python
# Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay không.
def point_in_quadrilateral(quad, p):
    def cross_product(o, a, b):
        return (a[0] - o[0]) * (b[1] - o[1]) - (a[1] - o[1]) * (b[0] - o[0])
    cp = []
    for i in range(4):
        cp.append(cross_product(quad[i], quad[(i + 1) % 4], p))    
    return all(x >= 0 for x in cp) or all(x <= 0 for x in cp)
quad_data = [(0,0), (4,0), (4,3), (0,3)]
p_data = (2, 1)
print(point_in_quadrilateral(quad_data, p_data))
```

```python
# Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)
def count_nonzero_digits(lst):
    count = 0
    for num in lst:
        for digit in str(abs(num)):
            if digit != '0':
                count += 1
    return count
n_list = list(map(int, input().split()))
print(count_nonzero_digits(n_list))
```
