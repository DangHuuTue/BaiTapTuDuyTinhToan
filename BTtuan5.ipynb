```python
#Viết hàm hoán đổi hai số nguyên.
def swap(a: int, b: int) -> Tuple[int, int]:
    return b, a
a, b = map(int, input().split())
swap(a, b)
```

```python
#Viết hàm kiểm tra một số có phải là số nguyên tố hay không?
def is_prime(n: int) -> bool:
    if n < 2: return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
n = int(input())
is_prime(n)
```

```python
#Viết hàm kiểm tra một số có phải số hoàn hảo hay không.
def is_perfect(n: int) -> bool:
    if n <= 1: return False
    sum_divisors = 1
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            sum_divisors += i
            if i*i != n:
                sum_divisors += n // i
    return sum_divisors == n
n = int(input())
is_perfect(n)
```

```python
#Tìm và trả về vị trí phần tử đầu tiên có giá trị k trong list, nếu không có trả về -1.
def find_first_k(lst: list, k: int) -> int:
    try:
        return lst.index(k)
    except ValueError:
        return -1
    
k = int(input())
lst = list(map(int, input().split()))
find_first_k(lst, k)
```

```python
#Viết hàm tính giai thừa n!
def factorial(n: int) -> int:
    if n < 0: return 0
    res = 1
    for i in range(2, n + 1):
        res *= i
    return res
n = int(input())
factorial(n)
```

```python
#Viết hàm giả lập một máy tính bỏ túi (+, -, *, /), làm tròn 2 chữ số thập phân.
def calculator(num1: float, num2: float, operator: str):
    if operator == '+': result = num1 + num2
    elif operator == '-': result = num1 - num2
    elif operator == '*': result = num1 * num2
    elif operator == '/': 
        result = num1 / num2 if num2 != 0 else "Error"
    else: return "Invalid"
    
    return round(result, 2) if isinstance(result, (int, float)) else result
num1, num2 = map(float, input().split())
operator = input().strip()
calculator(num1, num2, operator)
```

```python
#Viết hàm trả về khoảng cách Hamming giữa hai số nguyên x và y.
def hamming_distance(x: int, y: int) -> int:
    return bin(x ^ y).count('1')
x, y = map(int, input().split())
hamming_distance(x, y)
```

```python
#Viết hàm trả về tổng các chữ số của số nguyên dương n.
def sum_of_digits(n: int) -> int:
    return sum(int(digit) for digit in str(abs(n)))
n = int(input())
sum_of_digits(n)
```

```python
#Viết hàm kiểm tra hai từ a và b có đẳng cấu (isomorphic) hay không.
def is_isomorphic(a: str, b: str) -> bool:
    if len(a) != len(b): return False
    mapping = {}
    mapped_values = set()
    for char_a, char_b in zip(a, b):
        if char_a in mapping:
            if mapping[char_a] != char_b:
                return False
        else:
            if char_b in mapped_values:
                return False
            mapping[char_a] = char_b
            mapped_values.add(char_b)
    return True
a = input().strip()
b = input().strip()
is_isomorphic(a, b)
```

```python
#Trả về bộ (q, r) sao cho dividend = divisor * q + r.
def divide_with_remainder(dividend: int, divisor: int) -> Tuple[int, int]:
    if divisor == 0: return (0, 0)
    return divmod(dividend, divisor)
dividend = int(input())
divisor = int(input())
divide_with_remainder(dividend, divisor)
```

```python
#Viết hàm compute(a, b, op) thực hiện phép tính a (+ - * /) b.
def compute(a: float, b: float, op: str):
    if op == '+': return a + b
    if op == '-': return a - b
    if op == '*': return a * b
    if op == '/': return a / b if b != 0 else "Error"
    return None
a, b = map(float, input().split())
op = input().strip()
compute(a, b, op)
```

```python
#Viết hàm max_of_two(a, b) trả về số lớn hơn trong hai số.
def max_of_two(a, b):
    return max(a, b)
a, b = map(int, input().split())
max_of_two(a, b)
```

```python
#Viết hàm is_product_divisible_by_10(a, b) trả về True nếu tích a*b chia hết cho 10, ngược lại False.
def is_product_divisible_by_10(a, b):
    if (a * b) % 10 == 0:
        return True
    else:
        return False
a, b = map(int, input().split())
print(is_product_divisible_by_10(a, b))
```

```python
#Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông. 
def is_right_triangle(a, b, c):
    if a <= 0 or b <= 0 or c <= 0:
        return False
    else:
        if a**2 + b**2 == c**2 or a**2 + c**2 ==  b**2 or c**2 + b**2 == a**2:
            return True
        else:
            return False
a, b, c = map(int, input().split())
print(is_right_triangle(a, b, c))
```
