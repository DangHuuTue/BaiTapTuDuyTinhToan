```python
# Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng  giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả. Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]
nums = list(map(int, input().split()))
result = []
seen = set()
for x in nums:
    if x not in seen:
        result.append(x)
        seen.add(x)
print(result)
```

```python
# Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu tới vị trí i), giữ nguyên thứ tự
nums = list(map(int, input().split()))
res = []
current_sum = 0
for x in nums:
    current_sum += x
    res.append(current_sum)
print(res)
```

```python
# Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple sau khi xoay trái k vị trí.  (Giữ nguyên thứ tự tương đối sau xoay; không sắp xếp.)
t = tuple(map(int, input().split()))
k = int(input())
if len(t) == 0:
    print(())
else:
    k = k % len(t)
    res = t[k:] + t[:k]
    print(res)
```

```python
# Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện.  Nếu key lặp lại, thêm vào list của key đó
pairs = input().split()
d = {}
for p in pairs:
    key, val = p.split(':')
    if key not in d:
        d[key] = []
    d[key].append(val)
print(d)
```

```python
# Nhập một list số nguyên, In ra dictionary {'positives': ...., 'negatives': ..., 'zeros': ...} (đếm số >0, <0, =0)
nums = list(map(int, input().split()))
res = {'positives': 0, 'negatives': 0, 'zeros': 0}
for x in nums:
    if x > 0:
        res['positives'] += 1
    elif x < 0:
        res['negatives'] += 1
    else:
        res['zeros'] += 1
print(res)
```

```python
# Viết chương trình nhập vào một list các chuỗi, sau đó đếm số lần xuất hiện của từng chuỗi và lưu vào dictionary.
words = input().split()
d = {}
for w in words:
    d[w] = d.get(w, 0) + 1
print(d)
```

```python
# Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của key.
d1_input = input().split()
d2_input = input().split()
def parse_dict(inp):
    res = {}
    for item in inp:
        k, v = item.split(':')
        res[k] = int(v)
    return res
dict1 = parse_dict(d1_input)
dict2 = parse_dict(d2_input)
new_dict = dict1.copy()
for k, v in dict2.items():
    new_dict[k] = new_dict.get(k, 0) + v
sorted_dict = {k: new_dict[k] for k in sorted(new_dict.keys())}
print(sorted_dict)
```

```python
# Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là một tuple, thứ tự xuất hiện của số thứ nhất nhỏ hơn số thứ hai. Các tuple được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên.
nums = list(map(int, input().split()))
k = int(input())
res = []
n = len(nums)
for i in range(n):
    for j in range(i + 1, n):
        if nums[i] + nums[j] == k:
            res.append((nums[i], nums[j]))
res.sort()
print(res)
```

```python
# Nhập vào một tuple chứa các số nguyên. Hãy tạo ra hai tuple mới: một chứa tất cả các số chẵn và một chứa tất cả các số lẻ, giữ nguyên thứ tự.
t = tuple(map(int, input().split()))
chan = tuple(x for x in t if x % 2 == 0)
le = tuple(x for x in t if x % 2 != 0)
print(chan)
print(le)
```

```python
# Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất. Nếu có nhiều số cùng số lần, in ra số nhỏ nhất trong số đó.
nums = list(map(int, input().split()))
if not nums:
    pass
else:
    counts = {}
    for x in nums:
        counts[x] = counts.get(x, 0) + 1
    max_freq = max(counts.values())
    candidates = [k for k, v in counts.items() if v == max_freq]
    print(min(candidates))
```

```python
# Nhập vào một dictionary có value là duy nhất. Hãy tạo một dictionary mới bằng cách đảo ngược key và value.
items = input().split()
old_dict = {}
for item in items:
    k, v = item.split(':')
    old_dict[k] = v
new_dict = {v: k for k, v in old_dict.items()}
print(new_dict)
```

```python
# Nhập vào hai dòng, mỗi dòng là một danh sách số nguyên. Tìm các phần tử chung không trùng lặp và in ra list.
list1 = list(map(int, input().split()))
list2 = list(map(int, input().split()))
intersection = list(set(list1) & set(list2))
print(intersection)
```

```python
# Nhập vào một dictionary có key là chuỗi và value là số nguyên, và số nguyên k. Tạo dictionary mới chỉ chứa các cặp có value > k.
data = input().split()
k_limit = int(input())
old_dict = {}
for item in data:
    key, val = item.split(':')
    old_dict[key] = int(val)
new_dict = {k: v for k, v in old_dict.items() if v > k_limit}
print(new_dict)
```
